# What is Pandas?

Pandas is an open-source Python library for data manipulation and analysis. It gives you 
powerful, fast, flexible, and expressive data structures designed to work with structured data.

It is built on top of NumPy, and its name comes from â€œPanel Dataâ€ (econometrics term for multidimensional data).

# Why Pandas? What problem does it solve?

### Before Pandas:

People used raw Python lists, dictionaries, or NumPy arrays.

These structures lacked labels, indexing, grouping, join/merge, handling missing data, etc.

Real-world datasets (CSV, JSON, Excel, SQL) are structured in rows and columns â€” not arrays.

So Pandas was designed to work like:

ðŸ“Š Excel + ðŸ§® NumPy + ðŸ—‚ï¸ SQL + ðŸ§° ETL tools â€” inside Python

# 2. What is a DataFrame?

A DataFrame is a 2D labeled data structure with columns of potentially different types, just like a spreadsheet or SQL table.

Think of it like:

- A dict of Series

- Or a table with rows and columns

It is the core structure of pandas used for:

- Reading, cleaning, transforming, analyzing, and exporting data


### Why DataFrame?

Real-world data is usually tabular (e.g., CSV, Excel, SQL)

Rows = records

Columns = features

It supports label-based indexing, automatic alignment, missing data handling, grouping, aggregation, and more


| Concept     | Real-Life Analogy             |
| ----------- | ----------------------------- |
| `DataFrame` | Excel sheet or database table |
| `Series`    | One column of that sheet      |
| Index       | Row number or ID              |
| Columns     | Named headers                 |

# Are Subsets Views or Copies in Pandas?

In pandas, whether you get a view (linked to original DataFrame) or a copy (independent) 
depends on how you slice or filter â€” and itâ€™s not always obvious.

- .iloc always returns a copy (safe, independent subset.)
- .loc may return a view (especially if indexing on labels)
    ```bash
    subset = df.loc[df['Sex'] == 'male']
    This is like the earlier filtering â€” might be a view.

    so..
    subset = df.loc[df['Sex'] == 'male'].copy()

    ```

| Situation                          | View or Copy? | Should you use `.copy()`? |
| ---------------------------------- | ------------- | ------------------------- |
| `.iloc[]` slicing                  | Mostly Copy   | Optional                  |
| `.loc[]` slicing                   | May be View   | Recommended               |
| Filtering with condition `df[...]` | May be View   | Recommended               |
| You plan to modify the subset      | Use `.copy()` | YES                       |


````python
# instead of below

subset = df[df['Sex'] == 'male']
subset['Age'] = subset['Age'] + 1

# SettingWithCopyWarning: A value is trying to be set on a copy of a slice from a DataFrame in previous version
````

````python
subset = df[df['Sex'] == 'male'].copy()
subset['Age'] += 1
````

# 2. reindex() â€” Conform to a new index/column structure

````python
sales_jan = pd.DataFrame({
    'Sales': [250, 400, 300]
}, index=['Store_A', 'Store_B', 'Store_C'])

sales_jan
````

````python
sales_feb = pd.DataFrame({
    'Sales': [260, 310]
}, index=['Store_A', 'Store_C'])

sales_feb
````

````python
âŒ Problem: Indexes are mismatched! (above 2 dataframes)

You cannot safely do operations like subtraction or direct comparison unless indexes are aligned.

âœ… ðŸ›  Solution: Use reindex() to align!
````

````python
common_index = sales_jan.index

print(common_index)
````

````python
sales_feb_aligned = sales_feb.reindex(common_index)

sales_feb_aligned
````

````python
sales_diff = sales_feb_aligned['Sales'] - sales_jan['Sales']

sales_diff
````

````python
sales_feb_aligned = sales_feb.reindex(common_index, fill_value=0) # we use fill_value=0 to do correct calculations

sales_feb_aligned
````

````python
sales_diff = sales_feb_aligned['Sales'] - sales_jan['Sales']

sales_diff
````

# Iteration in Pandas

In pandas, iteration refers to looping over rows or columns of a DataFrame or Series. While pandas is optimized for vectorized operations, sometimes we need to loop â€” for example, for conditional logic, debugging, or generating complex values.

````python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['Delhi', 'Mumbai', 'Chennai']
})
````

````python
df
````

````python
for index, row in df.iterrows():
    print(f"{index}: {row['Name']} lives in {row['City']} and is {row['Age']} years old.")
````

````python
for row in df.itertuples():
    print(row)
    print(row.Index)
    print(f"{row.Name} lives in {row.City}, and is {row.Age} years old.", end='\n\n')
````

````python
# Iterating Over Columns

for col in df.columns:
    print(f"Column: {col}")
    print(df[col].values, end='\n\n')
````
