# Array Reshaping and resizing in Numpy**

### 1. reshape() - change shape without changing data.
  - The total no. of elements must remain the same.
  - reshape() returns a new view if possible (no copy).
  - you can use -1 to let Numpy infer one dimension.

```python
arr = np.array([1,2,3,4,5,6])
print(arr.shape)
```

```python
# 2*3
# 3*2
# 6*1
# 4*2 - Not possible
```

```python
reshaped_1 = arr.reshape(2,3)
print(reshaped_1)
```

```python
reshaped_1 = arr.reshape(3,2)
print(reshaped_1)
```

```python
reshaped_2 = arr.reshape(6,1)
print(reshaped_2)
```

```python
reshaped_3 = arr.reshape(1,6)
print(reshaped_3)
print(reshaped_3.ndim)
```

```python
arr_3 = np.array([1,2,3,4,5,6,34,55,77,33,56,77])
reshape4= arr_3.reshape(3,-1)  # here -1 means the reshape will auto decide the 2nd size by no of elements given in araay.
print(reshape4)
print(reshape4.shape)
print(reshape4.ndim)
```
