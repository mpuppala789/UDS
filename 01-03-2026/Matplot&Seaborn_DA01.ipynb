# Cleaning & Feature Engineering

```python
df.info()
```

```python
df=df.dropna(subset=['Customer ID'])  #dropping all customerID with null values the entire row with customerID is NULL
```

```python
df.info()
```

```python
df[100:200]  # we have some invoice with cancelled , need to remove cancelled invoice start with 'C'
```

```python
## Removing Cancellation Invoice ids, so the data should be precise
df = df[~df['Invoice'].astype(str).str.startswith('C')]
```

```python
df.shape
```

```python
824364-805620
```

```python
# Removing zero/negative quantities
df=df[df['Quantity']>0]
```

```python
df.shape
```

```python
df = df[df['Price'] > 0]
df.shape
```

```python
# Add new features

df['Revenue'] = df['Quantity'] * df['Price']
```

```python
df.info()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df['Month'] = df['InvoiceDate'].dt.to_period('M').astype(str)
```

# Seaborn: High-Level Statistical Data Visualization

Seaborn is a powerful Python visualization library built on top of Matplotlib and integrated with pandas. It simplifies many common tasks like:

- Statistical plots

- Categorical data plots

- Distribution plots

- Color themes

- Multi-variable analysis

```python
import seaborn as sns
```

```python
# New Analysis: Distribution of Quantity per Transaction

df_quantity = df[(df['Quantity'] > 0) & (df['Quantity'] < 100)]

# 1 to 99
```

```python
df_quantity.head()
```

#### Show me how transaction quantities (between 1 and 99) are distributed. Bin them into 10 groups, but also draw a smooth probability curve showing where quantity values tend to cluster.

```python
plt.figure(figsize=(12, 6))
sns.histplot(df_quantity['Quantity'], bins=10, kde=False, color='skyblue')
plt.title('Distribution of Revenue')
plt.xlabel('Quantity bins')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Kernel Density Estimate
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df_quantity['Quantity'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Revenue')
plt.xlabel('Quantity bins')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Kernel Density Estimate
```

```python
# sns.kdeplot() — KDE (smooth version of histogram)
plt.figure(figsize=(12, 8))
sns.kdeplot(df_quantity['Quantity'], fill=True, color='red')
plt.title('Revenue KDE Curve')
plt.xlabel('Revenue')
plt.tight_layout()
plt.show()
```
